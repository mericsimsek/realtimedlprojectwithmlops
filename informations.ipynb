Gördüğün Kısım (%10 - Kolay Olan): mlflow.start_run() yazmak, dvc add demek, bir Dockerfile yazmak. Bunlar araçları kullanmaktır.

Görmediğin Kısım (%90 - Asıl Zor Olan): MLOps'un zorluğu, bu araçları kullanarak otomatik, dayanıklı ve ölçeklenebilir bir sistem kurmaktır.

Zor olan kısmı "kodu yazmak" değil, "kodun kendi kendine yaşamasını sağlamaktır."

İşte MLOps'u "niş ve zor" yapan asıl 4 sebep:

1. Entegrasyon Cehennemi (Araçları Konuşturmak)
Senin projen basit ve tek bir kişi üzerinde. Şimdi şunu hayal et:

DVC'deki verilerin 5 TB'lık bir video arşivi ve Amazon S3'te duruyor.

MLflow sunucusu ayrı bir sanal makinede (VM) çalışıyor ve herkesin erişmesi lazım.

Docker imajı 30 GB büyüklüğünde çünkü içinde CUDA, PyTorch ve bir sürü karmaşık kütüphane var.

5 farklı makine öğrenmesi mühendisi aynı anda bu sistemi kullanmaya çalışıyor.

Zorluk: Bu 4-5 farklı servisin (S3, VM, Docker Registry, Git) birbiriyle sorunsuz, güvenli ve yetkili bir şekilde konuşmasını sağlamak. Birinin anahtarı (key) eksik olsa tüm sistem çöker.

2. Gerçek Otomasyon (CI/CD Pipeline'ları)
Şu an modeli eğitmek için terminale python train.py yazıyorsun.

Zorluk (Gerçek MLOps): Sen git push yaptığın an, GitHub Actions (veya Jenkins) gibi bir sistemin otomatik olarak şunları yapmasıdır:

Bir sanal makineyi tetiklemesi.

O makinede senin Dockerfile'ını "build" etmesi.

O Docker konteynerini çalıştırması.

Konteynerin içine dvc pull ile S3'ten doğru veri versiyonunu çekmesi.

python train.py komutunu çalıştırması (MLflow'a bağlanarak).

Eğitim bitince, MLflow'dan yeni modelin test accuracy'sini okuması.

Bu accuracy'nin, şu an canlıda (production) olan modelden daha iyi olup olmadığını kontrol etmesi.

Eğer daha iyiyse, modeli "Canlıya Alınabilir" olarak etiketlemesi.

Tüm bunları yaparken sana Slack'ten "Eğitim bitti, accuracy %95" diye mesaj atması.

İşte bu pipeline'ı yazmak ve hata ayıklamak zordur.

3. Ölçeklenebilir Dağıtım (Deployment)
Yarın Docker'ı bitirince elinde bir imaj olacak. Bunu docker run ... ile çalıştıracaksın.

Zorluk: Bu imajı bir API (FastAPI) ile sarmalayıp canlıya aldığını düşün. Peki ya aynı anda 10.000 kullanıcı senin işaret dili çevirmenini kullanmak isterse ne olacak?

Tek bir konteyner bu yükü kaldıramaz.

İşte burada Kubernetes (K8s) devreye girer.

Kubernetes'e "Bu imajdan al, yük artarsa otomatik olarak 50 tane çalıştır, yük azalınca 10'a düşür, biri çökerse anında yenisini başlat" demen gerekir.

Kubernetes'i kurmak, yönetmek ve ona bir modeli deploy etmek, MLOps'un en karmaşık kısımlarından biridir.

4. İzleme ve Geri Besleme Döngüsü (Monitoring & Drift)
Modeli canlıya aldın, her şey harika. Peki 6 ay sonra?

İnsanlar farklı ışık koşullarında, daha önce görmediğin yeni işaretler yapmaya başlarsa ne olur?

Modelinin accuracy'si canlıda yavaş yavaş düşmeye başlar. Buna "Model Drift" (Model Kayması) denir.

Zorluk: Canlıdaki modelin tahminlerini sürekli izleyen (Monitoring) bir sistem (Prometheus, Grafana gibi) kurmak. Bu sistemin, accuracy %90'ın altına düştüğü an otomatik olarak 2. maddedeki CI/CD pipeline'ını (yeniden eğitim) tetiklemesini sağlamak.
